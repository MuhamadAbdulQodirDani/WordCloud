# Menampilkan Data dalam Bentuk WorldCloud

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alamat = "/content/Copy of Data Statistik Channel YT Nabawi TV.xlsx"
dyoutube = pd.read_excel(alamat)
dyoutube.head(5)
```

# Membuat WorldCloud

```python
from wordcloud import WordCloud, STOPWORDS
comment_words = ''
stopwords = set(STOPWORDS)

# Iterate througth the csv file
for val in dyoutube["Unnamed: 0"]:

  #Typecaste each val to sttring
  val = str(val)

  # Split the value
  tokens = val.split()

  # Converts each token into lowercase
  for i in range(len(tokens)):
    tokens[i] = tokens[i].lower()

  comment_words += " ".join(tokens)+" "

worldcloud = WordCloud(width = 800, height = 800,
                       background_color = 'white',
                       stopwords = stopwords,
                       min_font_size = 10).generate(comment_words)

# Plot the WordCloud Image
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(worldcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
